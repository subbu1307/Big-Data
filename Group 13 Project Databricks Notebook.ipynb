# Categorical Features

```
#This code loops through each column, checks if the column is of string type and has 30 or fewer unique values, appends it to the object_col list, prints the unique values, and prints the value counts for each unique value. Finally, it removes the 'Attrition' column from the object_col list.

from pyspark.sql.functions import col, countDistinct

object_col = []

# Loop through each column and check for data type and unique values count
for column in df.columns:
    if df.select(column).dtypes[0][1] == 'string' and df.select(column).distinct().count() <= 30:
        object_col.append(column)
        unique_values = [row[column] for row in df.select(column).distinct().collect()]
        print(f"{column} : {unique_values}")
        value_counts = df.groupBy(column).count().orderBy('count', ascending=False)
        value_counts.show()
        print("====================================")

# Remove 'Attrition' from the list
object_col.remove('Attrition')
```

### 1. What is the attrition rate in the organization?

**📌 Business Insight: Helps determine the overall turnover percentage.**

```
%sql
SELECT 
    (COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0) / COUNT(*) AS attrition_rate
FROM hr_data;
```

### 2. Which department has the highest attrition rate?

**📌 Business Insight: Identifies departments needing targeted retention strategies.**

```
%sql
SELECT Department, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY Department
ORDER BY attrition_rate DESC;
```

### 3. How does job level impact attrition?

**📌 Business Insight: Lower job levels may need better career progression opportunities.**

```
%sql
SELECT JobLevel, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY JobLevel
ORDER BY JobLevel;
```

### 4. What is the impact of business travel on attrition?

**📌 Business Insight: Frequent travelers may need better work-life balance policies.**

```
%sql
SELECT BusinessTravel, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY BusinessTravel
ORDER BY attrition_rate DESC;
```

### 5. Does marital status affect attrition?

**📌 Business Insight: Singles might quit due to relocation flexibility—company could offer incentives.**

```
%sql
SELECT MaritalStatus, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY MaritalStatus
ORDER BY attrition_rate DESC;
```

### 6. How does work-life balance affect attrition?

**📌 Business Insight: Employees with poor work-life balance might need flexible schedules.**


```
%sql
SELECT WorkLifeBalance, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY WorkLifeBalance
ORDER BY WorkLifeBalance;
```

### 7. What is the average monthly income of employees who left vs. stayed?

**📌 Business Insight: If quitters have significantly lower salaries, a salary review strategy may be needed.**

```
%sql
SELECT Attrition, AVG(MonthlyIncome) AS avg_income
FROM hr_data
GROUP BY Attrition;
```

### 8. How does overtime impact attrition?

**📌 Business Insight: High overtime can lead to burnout and voluntary exits.**


```
%sql
SELECT OverTime, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY OverTime;
```

### 9. What job roles have the highest attrition?

**📌 Business Insight: Some job roles might need better retention incentives.**

```
%sql
SELECT JobRole, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY JobRole
ORDER BY attrition_rate DESC;
```

### 10. How does total working experience influence attrition?

**📌 Business Insight: Employees with fewer years of experience might need better career growth opportunities.**

```
%sql
SELECT TotalWorkingYears, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0 / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY TotalWorkingYears
ORDER BY TotalWorkingYears;
```

# A few complex questions to gain some depth !!

### 1. How does a combination of Job Role, OverTime, and Job Satisfaction impact attrition?

**📌 Business Insight: Employees in certain job roles with overtime work and low job satisfaction may be leaving at a higher rate. HR should focus on improving work-life balance & job satisfaction in those roles.**

```
%sql

-- This query helps analyze how overtime and job satisfaction affect employees in different job roles, giving a multi-dimensional insight.

SELECT JobRole, OverTime, JobSatisfaction, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) AS attrition_count,
       COUNT(*) AS total_employees,
       (COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0) / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY JobRole, OverTime, JobSatisfaction
ORDER BY attrition_rate DESC;
```

### 2. What is the correlation between Monthly Income, Years at Company, and Attrition?

**📌 Business Insight: If newer employees with lower salaries are leaving, it indicates that salary hikes or better onboarding programs might help in employee retention.**

```
%sql
--This query helps understand if low salaries combined with fewer years at the company contribute to higher attrition.
SELECT YearsAtCompany,
       ROUND(AVG(MonthlyIncome), 2) AS avg_income,
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) AS attrition_count,
       COUNT(*) AS total_employees,
       (COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0) / COUNT(*) AS attrition_rate
FROM hr_data
GROUP BY YearsAtCompany
ORDER BY YearsAtCompany;
```

### 3. Which business unit is losing its high performers?

**📌 Business Insight: If high-performing employees are leaving in large numbers from specific departments, HR must review leadership, promotions, and incentives to retain top talent.**


```
%sql
--This query identifies departments that are losing top-performing employees, which could indicate dissatisfaction among high-achieving employees.

SELECT Department, 
       PerformanceRating, 
       COUNT(CASE WHEN Attrition = 1 THEN 1 END) AS high_performer_attrition_count,
       COUNT(*) AS total_high_performers,
       (COUNT(CASE WHEN Attrition = 1 THEN 1 END) * 100.0) / COUNT(*) AS high_performer_attrition_rate
FROM hr_data
WHERE PerformanceRating >= 4
GROUP BY Department, PerformanceRating
ORDER BY high_performer_attrition_rate DESC;

```

##  Step 1: Import Required Libraries

```
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator, BinaryClassificationEvaluator
from pyspark.ml import Pipeline
from pyspark.sql.functions import col
import xgboost as xgb
from catboost import CatBoostClassifier
```

## Step 2: Encode Categorical Features & Assemble Features

```
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder

# Identify categorical columns
categorical_cols = [col_name for col_name, dtype in df.dtypes if dtype == 'string' and col_name != "Attrition"]

# String Indexer for categorical columns
indexers = [StringIndexer(inputCol=col, outputCol=col + "_Index").fit(df) for col in categorical_cols]

# OneHot Encoding (explicit feature naming)
encoders = [OneHotEncoder(inputCol=col + "_Index", outputCol=col + "_Vec") for col in categorical_cols]

# Assemble features (including numeric columns)
numeric_cols = [col for col, dtype in df.dtypes if dtype in ['int', 'double'] and col != "Attrition"]
feature_cols = [col + "_Vec" for col in categorical_cols] + numeric_cols
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
```

## Step 3: Split Data into Train & Test Sets

```
train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)
```

## Step 4: Logistic Regression Model

```
# Define Logistic Regression Model
log_reg = LogisticRegression(featuresCol="features", labelCol="Attrition", maxIter=10)

# Pipeline for Logistic Regression
log_pipeline = Pipeline(stages=indexers + encoders + [assembler, log_reg])

# Train Logistic Regression Model
log_model = log_pipeline.fit(train_df)

# Predictions
log_predictions = log_model.transform(test_df)
```

## Step 5: Random Forest Model

```
# Define Random Forest Model
rf = RandomForestClassifier(featuresCol="features", labelCol="Attrition", numTrees=100)

# Pipeline for Random Forest
rf_pipeline = Pipeline(stages=indexers + encoders + [assembler, rf])

# Train Random Forest Model
rf_model = rf_pipeline.fit(train_df)

# Predictions
rf_predictions = rf_model.transform(test_df)
```

## Step 6: XGBoost Model

```
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline
import numpy as np
import xgboost as xgb
from sklearn.metrics import roc_auc_score, average_precision_score

# Encode Categorical Features
categorical_cols = [col_name for col_name, dtype in df.dtypes if dtype == 'string' and col_name != "Attrition"]

# String Indexing
indexers = [StringIndexer(inputCol=col, outputCol=col + "_Index") for col in categorical_cols]

# One-Hot Encoding
encoders = [OneHotEncoder(inputCol=col + "_Index", outputCol=col + "_Vec") for col in categorical_cols]

# List of numeric features
numeric_cols = [col for col, dtype in df.dtypes if dtype in ['int', 'double'] and col != "Attrition"]

# Assemble all features into a single vector
feature_cols = [col + "_Vec" for col in categorical_cols] + numeric_cols
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

# Create and Fit Pipeline
pipeline = Pipeline(stages=indexers + encoders + [assembler])
pipeline_model = pipeline.fit(df)

# Transform train and test sets separately
train_transformed = pipeline_model.transform(train_df)
test_transformed = pipeline_model.transform(test_df)

# Debugging: Check if "features" column exists
train_transformed.printSchema()

# Convert Spark DataFrame to Pandas
train_pd = train_transformed.select("features", "Attrition").toPandas()
test_pd = test_transformed.select("features", "Attrition").toPandas()

# Extract feature vectors from SparseVector
X_train = np.array(train_pd["features"].apply(lambda x: x.toArray()).tolist())
X_test = np.array(test_pd["features"].apply(lambda x: x.toArray()).tolist())
y_train = train_pd["Attrition"]
y_test = test_pd["Attrition"]

# Train XGBoost Model
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1)
xgb_model.fit(X_train, y_train)

# XGBoost Predictions
xgb_predictions = xgb_model.predict(X_test)
xgb_probabilities = xgb_model.predict_proba(X_test)[:, 1]  # Probability for AUC
```

## Step 7: CatBoost Model

```
# Train CatBoost Classifier
cat_model = CatBoostClassifier(iterations=100, depth=6, learning_rate=0.1, verbose=0)
cat_model.fit(X_train, y_train)

# CatBoost Predictions
cat_predictions = cat_model.predict(X_test)
cat_probabilities = cat_model.predict_proba(X_test)[:, 1]  # Probability for AUC
```

## Step 8: Evaluate Model Performance


```
from sklearn.metrics import roc_auc_score, average_precision_score, accuracy_score

# Define Evaluators for Spark Models
accuracy_evaluator = MulticlassClassificationEvaluator(labelCol="Attrition", predictionCol="prediction", metricName="accuracy")
f1_evaluator = MulticlassClassificationEvaluator(labelCol="Attrition", predictionCol="prediction", metricName="f1")
auc_evaluator = BinaryClassificationEvaluator(labelCol="Attrition", rawPredictionCol="probability", metricName="areaUnderROC")
pr_evaluator = BinaryClassificationEvaluator(labelCol="Attrition", rawPredictionCol="probability", metricName="areaUnderPR")

# Compute Metrics for Spark Models
metrics = {
    "Logistic Regression": {
        "Accuracy": accuracy_evaluator.evaluate(log_predictions),
        "F1-Score": f1_evaluator.evaluate(log_predictions),
        "AUC (ROC)": auc_evaluator.evaluate(log_predictions),
        "Precision-Recall AUC": pr_evaluator.evaluate(log_predictions),
    },
    "Random Forest": {
        "Accuracy": accuracy_evaluator.evaluate(rf_predictions),
        "F1-Score": f1_evaluator.evaluate(rf_predictions),
        "AUC (ROC)": auc_evaluator.evaluate(rf_predictions),
        "Precision-Recall AUC": pr_evaluator.evaluate(rf_predictions),
    },
    "XGBoost": {
        "Accuracy": accuracy_score(y_test, xgb_predictions),
        "F1-Score": average_precision_score(y_test, xgb_predictions),
        "AUC (ROC)": roc_auc_score(y_test, xgb_probabilities),
        "Precision-Recall AUC": average_precision_score(y_test, xgb_probabilities),
    },
    "CatBoost": {
        "Accuracy": accuracy_score(y_test, cat_predictions),
        "F1-Score": average_precision_score(y_test, cat_predictions),
        "AUC (ROC)": roc_auc_score(y_test, cat_probabilities),
        "Precision-Recall AUC": average_precision_score(y_test, cat_probabilities),
    },
}

# Print Results
for model, scores in metrics.items():
    print(f"📊 {model} Model Performance:")
    print(f"   ✅ Accuracy: {scores['Accuracy']:.4f}")
    print(f"   ⚡ F1-Score: {scores['F1-Score']:.4f}")
    print(f"   🔥 AUC (ROC): {scores['AUC (ROC)']:.4f}")
    print(f"   🎯 Precision-Recall AUC: {scores['Precision-Recall AUC']:.4f}")
    print("-" * 40)
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier

# ✅ Encode Categorical Features
categorical_cols = [col for col, dtype in df.dtypes if dtype == 'string' and col != "Attrition"]
indexers = [StringIndexer(inputCol=col, outputCol=col + "_Index") for col in categorical_cols]
encoders = [OneHotEncoder(inputCol=col + "_Index", outputCol=col + "_Vec") for col in categorical_cols]

# ✅ Assemble Features
feature_cols = [col + "_Vec" for col in categorical_cols] + [col for col, dtype in df.dtypes if dtype in ['int', 'double'] and col != "Attrition"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

# ✅ Define Models
log_reg = LogisticRegression(featuresCol="features", labelCol="Attrition", maxIter=10)
rf = RandomForestClassifier(featuresCol="features", labelCol="Attrition", numTrees=100)

# ✅ Create Pipelines for Both Models
log_pipeline = Pipeline(stages=indexers + encoders + [assembler, log_reg])
rf_pipeline = Pipeline(stages=indexers + encoders + [assembler, rf])

# ✅ Split Data
train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)
```
